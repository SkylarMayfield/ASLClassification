```python
# Imorting packages
import cv2
import csv
from google.colab.patches import cv2_imshow
import mediapipe as mp
import os

import pandas as pd
from google.colab import drive
```

```python
# This function tracks the frame number and displays information (coordinates x,y,z) about the various
# poses modeled in each frame. Each tracked position is then added to a data csv file.
def track_position_coords(position, frame_num, csv_data):
    # Prints out the corresponding list of body positoins by frame
    #print(f"Positions corresponding to {frame_num}:")

    # Appending frames numbers to the frame list
    frame = []
    frame.append(frame_num)
    i = 0
    # Iterating through the body position and calculating the coordinates
    for index, pos in enumerate(position):
        #print(f"{mp_pose.PoseLandmark(index).name}: (x: {pos.x}, y: {pos.y}, z: {pos.z})")
        frame.append(pos.x)
        frame.append(pos.y)
        frame.append(pos.z)
    #print("\n")

    # Adding the frame numbers to the csv file
    csv_data.append(frame)
```

```python
# Creating a path to the video of the sign language and a path to where the output
# csv should be output
folder_path = '/content/drive/MyDrive/MLProject/book-drink-videos/book_and_drink' #  path of the video folder
output_csv = '/content/drive/MyDrive/MLProject/book-drink-csv' # path for output folder
```

```python
# Preparing the MediaPipe Pose and drawing_utils in order to track and draw
# the movements and positions
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils
pose = mp_pose.Pose()

# Opening the video file that contains the sign langauge
cap = cv2.VideoCapture(folder_path)
```

```python
# This block of code prepares and analyzes the video and tracks the position/body movements that are being signed (along with
# the positions of the facial features). The corresponding frame is then printed out along with each of the coordinates for the
# body positions, and the data is formatted into the csv file

# Iterating through each video in the folder
for video_file in os.listdir(folder_path):
    video_path = os.path.join(folder_path, video_file)
    cap = cv2.VideoCapture(video_path)

    print(video_file)

    # Frame number initlized to 0
    frame_number = 0

    # List created to hold csv informtaion about frame numbers and body positions
    csv_data = []

    # Formatting for csv file created
    labels = ["frame_number", "nose_x", "nose_y", "nose_z", "left_eye_inner_x", "left_eye_inner_y", "left_eye_inner_z", "left_eye_x", "left_eye_y", "left_eye_z", "left_eye_outer_x", "left_eye_outer_y", "left_eye_outer_z", "right_eye_inner_x", "right_eye_inner_y", "right_eye_inner_z", "right_eye_x", "right_eye_y", "right_eye_z", "right_eye_outer_x", "right_eye_outer_y", "right_eye_outer_z", "left_ear_x", "left_ear_y", "left_ear_z", "right_ear_x", "right_ear_y", "right_ear_z", "mouth_left_x", "mouth_left_y", "mouth_left_z", "mouth_right_x", "mouth_right_y", "mouth_right_z", "left_shoulder_x", "left_shoulder_y", "left_shoulder_z", "right_shoulder_x", "right_shoulder_y", "right_shoulder_z", "left_elbow_x", "left_elbow_y", "left_elbow_z", "right_elbow_x", "right_elbow_y", "right_elbow_z", "left_wrist_x", "left_wrist_y", "left_wrist_z", "right_wrist_x", "right_wrist_y", "right_wrist_z", "left_pinky_x", "left_pinky_y", "left_pinky_z", "right_pinky_x", "right_pinky_y", "right_pinky_z", "left_index_x", "left_index_y", "left_index_z", "right_index_x", "right_index_y", "right_index_z", "left_thumb_x", "left_thumb_y", "left_thumb_z", "right_thumb_x", "right_thumb_y", "right_thumb_z", "left_hip_x", "left_hip_y", "left_hip_z", "right_hip_x", "right_hip_y", "right_hip_z", "left_knee_x", "left_knee_y", "left_knee_z", "right_knee_x", "right_knee_y", "right_knee_z", "left_ankle_x", "left_ankle_y", "left_ankle_z", "right_ankle_x", "right_ankle_y", "right_ankle_z", "left_heel_x", "left_heel_y", "left_heel_z", "right_heel_x", "right_heel_y", "right_heel_z", "left_foot_index_x", "left_foot_index_y", "left_foot_index_z", "right_foot_index_x", "right_foot_index_y", "right_foot_index_z"
    ]
    csv_data.append(labels)

    # While the video object is open
    while cap.isOpened():
        # Saves values ret (boolean of if the video has been read) and the frame of the video
        read, frame = cap.read()

        # If the video is no longer being read, we break
        if not read:
            break

        # Converts the frame to RGB
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = pose.process(frame_rgb)

        # Frame is being processed
        video = pose.process(frame_rgb)

        # The features/positions are being drawn on the frame
        # if video.pose_landmarks:
        #     mp_drawing.draw_landmarks(frame, video.pose_landmarks, mp_pose.POSE_CONNECTIONS)

        # The coordinates are being calculated and added to the csv data list
         # Check if landmarks were detected
        if results.pose_landmarks is not None:
            # Landmarks detected, process them
            track_position_coords(results.pose_landmarks.landmark, frame_number, csv_data)
        else:
            # No landmarks detected, frames are skipped
            print(f"No landmarks detected in frame {frame_number}")

        # The current frame is displayed
        #cv2_imshow(frame)
        frame_number += 1

    # The data list is being saved as a pandas dataframe, and then formatted as a csv file
    # closing the videos
    cap.release()


    # writing to output csv
    output_csv_path = os.path.join(output_csv, f"{os.path.splitext(video_file)[0]}.csv")
    with open(output_csv_path, 'w', newline='') as f:
            writer = csv.writer(f)
            writer.writerows(csv_data)

cv2.destroyAllWindows()
```
